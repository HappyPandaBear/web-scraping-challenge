```python
from bs4 import BeautifulSoup as bs
from splinter import Browser
import pandas as pd
from webdriver_manager.chrome import ChromeDriverManager
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://redplanetscience.com/'
browser.visit(url)
```

```python
html = browser.html
soup = bs(html, 'html.parser')

print(soup.prettify())
```

```python
title = soup.find("div", class_ = "content_title").text
paragraph = soup.find("div", class_ = "article_teaser_body").text

print(title)
print(paragraph)
```

```python
space_url = 'https://spaceimages-mars.com/'
browser.visit(space_url)

image_html = browser.html
soup = bs(image_html, 'html.parser')

image_url = soup.find_all('img')[1]["src"]
space_url = 'https://spaceimages-mars.com/'

featured_image_url = space_url + image_url
featured_image_url
```

```python
hemispheres_url = "https://marshemispheres.com/"
browser.visit(hemispheres_url)

html_hemispheres = browser.html
soup = bs(html_hemispheres, 'html.parser')
images = soup.find_all('div', class_='item')

hemispheres_img_url = []
hemispheres_main_url = "https://marshemispheres.com/"

for image in images: 
    title = image.find('h3').text
    partial_img_url = image.find('a', class_='itemLink product-item')['href']
    browser.visit(hemispheres_main_url + partial_img_url)
    partial_img_html = browser.html
    soup = bs(partial_img_html, 'html.parser')
    img_url = hemispheres_main_url + soup.find('img', class_='wide-image')['src']
    hemispheres_img_url.append({"title" : title, "img_url" : img_url})
    
hemispheres_img_url
```
